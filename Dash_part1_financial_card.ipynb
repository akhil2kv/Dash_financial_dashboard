```python
import dash
import dash_core_components as dcc
import dash_html_components as dtc
from dash.dependencies import Input,Output
import dash_bootstrap_components as dbc
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from alpha_vantage.timeseries import TimeSeries
```

```python
key = '920CRW9UQMJBPZHM'
ts = TimeSeries(key=key,output_format="pandas")
ttm_data ,ttm_meta_data = ts.get_intraday(symbol='TTM',interval='1min',outputsize='compact')
ttm_data.head(5)
```

```python
df = ttm_data.copy( ) #make copy of the DaraFrame so that original will be preserved this way
df = df.transpose() # make a transpose of the DataFrame for better working 
df.head()
```

```python
df.rename(index={"1. open":"open", "2. high":"high", "3. low":"low",
                 "4. close":"close","5. volume":"volume"},inplace=True) # just renaming to proper formats

df=df.reset_index() # converting multi-index into standard index format 
df=df.rename(columns={'index':'indicator'}) # renaming index column
df.head(5)
```

```python
df = pd.melt(df,id_vars='indicator',var_name='date',value_name='rate')
df.head()
```

```python
df.info()
```

```python
df = df[df['indicator']!= 'volume']
```

```python
px.scatter(df,x='date',y='rate',color='indicator')
```
